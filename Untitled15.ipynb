```python
from osgeo import gdal
import numpy as np
import matplotlib.pyplot as plt
def scaleMinMax(x):
  return((x-np.nanmin(x))/(np.nanmax(x)-np.nanmin(x)))
def scaleCCC(x):
  return((x-np.nanpercentile(x,2))/(np.nanpercentile(x,98)-np.nanpercentile(x,2)))

ds=gdal.Open("/content/S2A_MSIL1C_20220516_TrainingData.tif")

r =ds.GetRasterBand(2).ReadAsArray()
g =ds.GetRasterBand(4).ReadAsArray()
b =ds.GetRasterBand(3).ReadAsArray()
r=scaleMinMax(r)
g=scaleMinMax(g)
b=scaleMinMax(b)

ds=None

rgb= np.dstack((r,g,b))
plt.figure()
plt.imshow(rgb)
plt.show()
plt.hist(r.flatten(),bins=50) 










```
